```python
from sklearn import preprocessing
import pandas as pd
import numpy as np
import sys
import torch.optim as optim
import torch
import torch.nn as nn
from d2l import torch as d2l
import matplotlib.pyplot as plt
import numpy as np
from torch.utils.data import Dataset, DataLoader, TensorDataset
from sklearn.cluster import KMeans
from sklearn import preprocessing
```

```python
##导入数据
data = pd.read_csv('E:\实验数据\python\May_data\data.csv')
data = data.values

##转换成array
data_array = data.astype('double')

train_x = data_array[:, :-1]
train_y = data_array[:, [-1]]
```

```python
data
```

```python
##归一化
x_scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))
y_scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))

# train_X = x_scaler.fit_transform(train_x)
train_X = train_x
train_Y = y_scaler.fit_transform(train_y)

# pp = scaler.inverse_transform(data_array)
```

```python
#转换成tensor
input_total = torch.tensor(train_X,dtype=torch.float32)     #只经过归一化的原始数据集
height_labels = torch.tensor(train_Y,dtype=torch.float32)
```

```python
batch_size, n_train = 16, 1200
train_iter = d2l.load_array((input_total[:n_train], height_labels[:n_train]), batch_size, is_train=True)
def init_weights(m):
    if type(m) == nn.Linear:
        nn.init.xavier_uniform_(m.weight)

def get_net():
    net = nn.Sequential(nn.Linear(6, 50),nn.ReLU(),nn.Linear(50, 50),nn.ReLU(),nn.Linear(50, 1))
    net.apply(init_weights)
    return net

loss = nn.MSELoss()
```

```python
def train(net, train_iter, loss, epochs, lr):
    trainer = torch.optim.Adam(net.parameters(), lr)
    plot_loss = torch.zeros(epochs)
    for epoch in range(epochs):
        for X, y in train_iter:
            trainer.zero_grad()
            l = loss(net(X.float()), y.float())
            l.backward()
            trainer.step()
            
        print(f'epoch {epoch + 1}, '
              f'loss: {d2l.evaluate_loss(net, train_iter, loss):f}')
        
        plot_loss[epoch] = d2l.evaluate_loss(net, train_iter, loss)
    plot_loss_csv = pd.DataFrame({"loss":plot_loss[:]});
    plot_loss_csv.plot(figsize=(15,10), title="Lo1ss-Epochs", color='green', linestyle='-', xlabel='epoch', ylabel='loss')

#超参数设置
net = get_net()
train_times = 1000
lr = 0.005
train(net, train_iter, loss, train_times, lr)
```

```python
plt.figure(figsize=(20, 10))
y = net(input_total)   #用所有数据进行检验
show_labels = y_scaler.inverse_transform(height_labels)  #反归一化
y = y_scaler.inverse_transform(y.detach().numpy())

plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.plot(show_labels, linestyle='-', color='red', linewidth=2, label='source')
plt.plot(y, linestyle='-', color='blue', linewidth=2, label='predict')
# plt.legend(fontsize=20)

error = y - show_labels
rmse = np.mean(error ** 2) ** 0.5
print("rmse:", rmse.item())
```
