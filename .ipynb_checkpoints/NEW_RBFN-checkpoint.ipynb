```python
import os
import pandas as pd
import numpy as np
import torch, random
import torch.nn as nn
from torch.utils import data
import torch.optim as optim
import matplotlib.pyplot as plt
from d2l import torch as d2l
```

```python
torch.manual_seed(42)
 
class RBFN(nn.Module):
    """
    以高斯核作为径向基函数
    """
    def __init__(self, centers, n_out=3):
        """
        :param centers: shape=[center_num,data_dim]
        :param n_out:
        """
        super(RBFN, self).__init__()
        self.n_out = n_out
        self.num_centers = centers.size(0) # 隐层节点的个数
        self.dim_centure = centers.size(1) # 
        self.centers = nn.Parameter(centers)
        # self.beta = nn.Parameter(torch.ones(1, self.num_centers), requires_grad=True)
        self.beta = torch.ones(1, self.num_centers)*10
        # 对线性层的输入节点数目进行了修改
        self.linear = nn.Linear(self.num_centers+self.dim_centure, self.n_out, bias=True)
        self.initialize_weights()# 创建对象时自动执行
 
 
    def kernel_fun(self, batches):
        n_input = batches.size(0)  # number of inputs
        A = self.centers.view(self.num_centers, -1).repeat(n_input, 1, 1)
        B = batches.view(n_input, -1).unsqueeze(1).repeat(1, self.num_centers, 1)
        C = torch.exp(-self.beta.mul((A - B).pow(2).sum(2, keepdim=False)))
        return C
 
    def forward(self, batches):
        radial_val = self.kernel_fun(batches)
        class_score = self.linear(torch.cat([batches, radial_val], dim=1))
        return class_score
 
    def initialize_weights(self, ):
        """
        网络权重初始化
        :return:
        """
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                m.weight.data.normal_(0, 0.02)
                m.bias.data.zero_()
            elif isinstance(m, nn.ConvTranspose2d):
                m.weight.data.normal_(0, 0.02)
                m.bias.data.zero_()
            elif isinstance(m, nn.Linear):
                m.weight.data.normal_(0, 0.02)
                m.bias.data.zero_()
 
    def print_network(self):
        num_params = 0
        for param in self.parameters():
            num_params += param.numel()
        print(self)
        print('Total number of parameters: %d' % num_params)
        

# centers = torch.rand((5,8))
# rbf_net = RBFN(centers)
# rbf_net.print_network()
# rbf_net.initialize_weights()
```

```python
input_file_name = "E:/实验数据/csv/train_input.csv"
input_data = pd.read_csv(input_file_name)

height_file_name = "E:/实验数据/csv/train_data.csv"
height_data_csv = pd.read_csv(height_file_name);


################################# 减最小值除最大值的归一化方法 ##############################
input_min = np.zeros(input_data.shape[0])
input_max = np.zeros(input_data.shape[0])
height_minv = np.min(np.min(height_data_csv))
height_maxv = np.max(np.max(height_data_csv))
for i in range(0,54):
    input_min[i] = np.min(input_data.iloc[i])
    input_max[i] = np.max(input_data.iloc[i])
    input_data.iloc[i] = (input_data.iloc[i] - input_min[i]) / input_max[i]
    
height_data_csv = (height_data_csv - height_minv) / (height_maxv - height_minv)

input_total = torch.tensor([], dtype=torch.float32)
labels = torch.tensor([], dtype=torch.float32)

############################################
### 生成序列数据
### tau    表示输入序列的样本个数 
### h_tau  表示高度序列的样本个数
def get_data_seq(input_1, input_2, height, tau, h_tau):

    T = input_1.shape[0];
    features_u1 = torch.zeros((T - tau), tau, dtype=torch.float32)
    features_u2 = torch.zeros((T - tau), tau, dtype=torch.float32)
    my_height = torch.zeros((T - tau), tau, dtype=torch.float32)
    res_height = torch.zeros((T - tau), h_tau, dtype=torch.float32)
    for i in range(tau):
        features_u1[:, i] = input_1[i:T - tau + i]
        features_u2[:, i] = input_2[i:T - tau + i]
        my_height[:, i] = height[i:T - tau + i]
    local_labels = height[tau:].reshape((-1, 1))
    
    
    for j in range(1, h_tau+1):
        for i in range(T - tau):
            res_height[i,j-1] = my_height[i,tau-j-1]
    

    return (torch.cat((features_u1, features_u2, res_height), 1), local_labels)
```

```python
U_TAU = 3
H_TAU = 2

for i in range(1, 28):
    
    # data.iloc[9, :].plot()
    experment = i - 1
    input_data_u1 = input_data.iloc[2*experment, :]
    input_data_u2 = input_data.iloc[2*experment + 1, :]
    height_data_pd = height_data_csv.iloc[experment,:]

    ########################################
    my_input_u1 = np.array(input_data_u1)
    my_tensor_input_u1 = torch.tensor(my_input_u1, dtype=torch.float32)

    my_input_u2 = np.array(input_data_u2)
    my_tensor_input_u2 = torch.tensor(my_input_u2, dtype=torch.float32)

    my_height = np.array(height_data_pd)
    my_tensor_height = torch.tensor(my_height, dtype=torch.float32)

    input_u, labels_u = get_data_seq(my_tensor_input_u1, my_tensor_input_u2, my_tensor_height, U_TAU, H_TAU)

    input_total = torch.cat((input_total, input_u), dim=0)
    labels = torch.cat((labels, labels_u), dim=0)
```

```python
centers = input_total[0:15,:]
centers
```

```python
horizen = 1  ##向后预测1个点
rbf = RBFN(centers,1) 
params = rbf.parameters()
loss_fn = torch.nn.MSELoss()
optimizer = torch.optim.SGD(params,lr=0.1,momentum=0.9)
```

```python
for i in range(10000):
        optimizer.zero_grad()
 
        y = rbf.forward(input_total.data)
        loss = loss_fn(y,labels)
        loss.backward()
        optimizer.step()
#         print(i,"\t",loss.data)
        
# 加载使用
y = rbf.forward(input_total)

labels = labels *(height_maxv - height_minv) + height_minv
y = y * (height_maxv - height_minv) + height_minv

error = y - labels
rmse = torch.mean(error ** 2) ** 0.5
print("rmse: ", rmse)
print("error: ", error)
```
