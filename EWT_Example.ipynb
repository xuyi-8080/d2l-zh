```python
import pywt
import numpy as np
import pandas as pd
from scipy.interpolate import make_interp_spline
import matplotlib.pyplot as plt
import ewtpy
```

```python
df = pd.read_csv('E:\实验数据\python\LSTM_data\denoise_data.csv', encoding='gbk')
load = df[df.columns[2]]
# load = load.tolist()
```

```python
ewt,mfb,boundaries=ewtpy.EWT1D(load,N=3)
```

```python
a1 = []
a2 = []
a3 = []
for i in range(len(ewt)):
    a1.append(ewt[i][0])
    a2.append(ewt[i][1])
    a3.append(ewt[i][2])
```

```python
df1 = pd.DataFrame(data=ewt, columns=['label_1','label_2','label_3'])
df1.to_csv('E:\实验数据\python\LSTM_data\wavelet_feature.csv',index=False)
```

```python
plt.figure(figsize=(8, 5), dpi=100)
plt.subplot(4,1,1)
plt.plot(load)

plt.subplot(4,1,2)
plt.plot(a1)

plt.subplot(4,1,3)
plt.plot(a2)

plt.subplot(4,1,4)
plt.plot(a3)
```

```python
data=pd.DataFrame({'a1':a1,'a2':a2, 'a3':a3, 'h':load})
corr=data.corr(method='pearson')
print(corr)
```
