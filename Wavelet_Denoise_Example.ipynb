```python
import numpy as np
import pywt
import pandas as pd
from skimage.restoration import denoise_wavelet
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('E:\实验数据\python\LSTM_data\data.csv', encoding='gbk')
load = df[df.columns[2]]
# load = load.tolist()
```

```python
x_denoise = denoise_wavelet(load, method='BayesShrink', mode='soft',wavelet_levels=8,wavelet='db1',rescale_sigma='True')
```

```python
plt.figure(figsize=(8, 5), dpi=100)
plt.plot(load)
plt.plot(x_denoise)
```

```python
df1 = pd.DataFrame(data=x_denoise, columns=['label'])
df1.to_csv('E:\实验数据\python\LSTM_data\wavelet.csv',index=False)
```
