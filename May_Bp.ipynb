```python
from sklearn import preprocessing
import pandas as pd
import numpy as np
import sys
import torch.optim as optim
import torch
import torch.nn as nn
from d2l import torch as d2l
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
import time
```

```python
##导入数据
data = pd.read_csv('E:\实验数据\python\RNN_data\data_1.csv')
data = data.values

##转换成array
data_array = data.astype('double')

train_x = data_array[:, :-1]
train_y = data_array[:, [-1]]
```

```python
##归一化
x_scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))
y_scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))

# train_X = train_x
# train_Y = train_y

train_X = x_scaler.fit_transform(train_x)
train_Y = y_scaler.fit_transform(train_y)
```

```python
#转换成tensor
input_total = torch.tensor(train_X,dtype=torch.float32)     #只经过归一化的原始数据集
height_labels = torch.tensor(train_Y,dtype=torch.float32)
```

```python
batch_size, n_train = 16, 1200
train_iter = d2l.load_array((input_total[:n_train], height_labels[:n_train]), batch_size, is_train=True)
def init_weights(m):
    if type(m) == nn.Linear:
        nn.init.xavier_uniform_(m.weight)

def get_net():
    net = nn.Sequential(nn.Linear(6, 50),nn.ReLU(),nn.Linear(50, 1))
    net.apply(init_weights)
    return net

loss = nn.MSELoss()
```

```python
def train(net, train_iter, loss, epochs, lr):
    trainer = torch.optim.SGD(net.parameters(), lr)
    plot_loss = torch.zeros(epochs)
    for epoch in range(epochs):
        for X, y in train_iter:
            l = loss(net(X.float()), y.float())
            trainer.zero_grad()
            l.backward()
            trainer.step()
            
        print(f'epoch {epoch + 1}, '
              f'loss: {d2l.evaluate_loss(net, train_iter, loss):f}')
        
        plot_loss[epoch] = d2l.evaluate_loss(net, train_iter, loss)
    plt.figure(figsize=(10, 5))
    plot_loss_csv = pd.DataFrame({"loss":plot_loss[:]});
    plot_loss_csv.plot(figsize=(15,10), title="Lo1ss-Epochs", color='green', linestyle='-', xlabel='epoch', ylabel='loss')

#超参数设置
net = get_net()
train_times = 5000
lr = 0.01
t1 = time.time()
train(net, train_iter, loss, train_times, lr)
train_time = time.time() - t1
```
