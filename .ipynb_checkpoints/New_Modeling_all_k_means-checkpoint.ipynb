```python
import os
import pandas as pd
import numpy as np
import torch
from torch.utils import data
from d2l import torch as d2l
from torch import nn  #nn中有大量定义好的神经网络的层
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
input_file_name = "E:/实验数据/P_train.csv"
input_data = pd.read_csv(input_file_name)

height_file_name = "E:/实验数据/P_height.csv"
height_data = pd.read_csv(height_file_name)

min_input = np.zeros(4)
max_input = np.zeros(4)

################################ 减最小值除最大值的归一化方法 ##############################
for i in range(input_data.shape[1]):   #分别计算每一列的最大值最小值, 分别进行归一化
    minv = np.min(input_data.iloc[:,i])
    maxv = np.max(input_data.iloc[:,i])
    input_data.iloc[:,i] = (input_data.iloc[:,i] - minv) / (maxv - minv)
    min_input[i] = minv
    max_input[i] = maxv

input_data = np.array(input_data)
    
height_minv = np.min(height_data).iloc[0]
height_maxv = np.max(height_data).iloc[0]
height_data = (height_data - height_minv) / (height_maxv - height_minv)

height_data = np.array(height_data)

input_total = torch.tensor(input_data,dtype=torch.float32)     #只经过归一化的原始数据集
height_labels = torch.tensor(height_data,dtype=torch.float32)
```

```python
k = 4
kmodel = KMeans(n_clusters = k).fit(input_data)   ##对输入数据进行 K-Means聚类
# kmodel.cluster_centers_    ##这个存放了结果的中心点
# print(kmodel.labels_ )     ##这个存放了每个点的分类
```

```python
#计算两个向量的欧式距离, 输入一定要是同纬度的 np.array 类型的向量  
def distance(vector_A, vector_B):
    dis = vector_A - vector_B
    dis = np.sum((dis**2))**0.5
    return dis
```

```python
##@算一下每一类点距离中心点的均值, 设定一个边界距离, 大于这个距离的点都筛掉


res = np.zeros(4)  ##存放每个类的数据点距离中心点的距离之和
cnt = np.zeros(4)  ##存放每个类的数据点的个数
for i in range(height_data.shape[0]):
    if kmodel.labels_[i] == 0:
        res[0] = res[0] + distance(kmodel.cluster_centers_[0], input_data[i])
        cnt[0] = cnt[0] + 1
    elif kmodel.labels_[i] == 1:
        res[1] = res[1] + distance(kmodel.cluster_centers_[1], input_data[i])
        cnt[1] = cnt[1] + 1
    elif kmodel.labels_[i] == 2:
        res[2] = res[2] + distance(kmodel.cluster_centers_[2], input_data[i])
        cnt[2] = cnt[2] + 1
    elif kmodel.labels_[i] == 3:
        res[3] = res[3] + distance(kmodel.cluster_centers_[3], input_data[i])
        cnt[3] = cnt[3] + 1

boundary_distance = res / cnt   ## 计算边界距离
print(boundary_distance)

k_means_data = []    ##用来保存经过筛除后的输入数据, 定义成list方便添加元素, 之后转换成np.array
k_means_height = []  ##用来保存经过筛除后的高度
##筛点
for i in range(height_data.shape[0]):
    if kmodel.labels_[i] == 0:
        if (distance(input_data[i], kmodel.cluster_centers_[0])) < (boundary_distance[0]):
            k_means_data.append(input_data[i])
            k_means_height.append(height_data[i])
    if kmodel.labels_[i] == 1:
        if (distance(input_data[i], kmodel.cluster_centers_[1])) < (boundary_distance[1]): 
            k_means_data.append(input_data[i])
            k_means_height.append(height_data[i])
    if kmodel.labels_[i] == 2:
        if (distance(input_data[i], kmodel.cluster_centers_[2])) < (boundary_distance[2]):
            k_means_data.append(input_data[i])
            k_means_height.append(height_data[i])
    if kmodel.labels_[i] == 3:
        if (distance(input_data[i], kmodel.cluster_centers_[3])) < (boundary_distance[3]):
            k_means_data.append(input_data[i])
            k_means_height.append(height_data[i])

k_means_data = np.array(k_means_data)
k_means_height = np.array(k_means_height)

##反归一化然后输出为csv文件保存
pf = pd.DataFrame(k_means_data)
pf.to_csv("E:/实验数据/test1.csv", encoding="utf-8-sig", header=False, index=False)
pf = pd.DataFrame(k_means_height)
pf.to_csv("E:/实验数据/test2.csv", encoding="utf-8-sig", header=False, index=False)

k_means_data = torch.tensor(k_means_data, dtype=torch.float32)    ##将经过筛除的数据转换成tensor 准备打包成数据迭代器, 送入神经网络
k_means_height = torch.tensor(k_means_height, dtype=torch.float32)
```

```python
batch_size, n_train = 16, 700
train_iter = d2l.load_array((k_means_data[:n_train], k_means_height[:n_train]), batch_size, is_train=True)
def init_weights(m):
    if type(m) == nn.Linear:
        nn.init.xavier_uniform_(m.weight)

def get_net():
    net = nn.Sequential(nn.Linear(4, 20),nn.ReLU(), nn.Linear(20, 1))
    net.apply(init_weights)
    return net

loss = nn.MSELoss()
```

```python
def train(net, train_iter, loss, epochs, lr):
    trainer = torch.optim.Adam(net.parameters(), lr)
    plot_loss = torch.zeros(epochs)
    for epoch in range(epochs):
        for X, y in train_iter:
            trainer.zero_grad()
            l = loss(net(X.float()), y.float())
            l.backward()
            trainer.step()
            
        print(f'epoch {epoch + 1}, '
              f'loss: {d2l.evaluate_loss(net, train_iter, loss):f}')
        
        plot_loss[epoch] = d2l.evaluate_loss(net, train_iter, loss)
    plot_loss_csv = pd.DataFrame({"loss":plot_loss[:]});
    plot_loss_csv.plot(figsize=(15,10), title="Lo1ss-Epochs", color='green', linestyle='-', xlabel='epoch', ylabel='loss')

net = get_net()
train_times = 1000
lr = 0.005
train(net, train_iter, loss, train_times, lr)
```

```python
plt.figure(figsize=(20, 10))
y = torch.tensor([])
y = net(input_total)   #用所有数据进行检验
show_labels = height_labels*(height_maxv - height_minv) + height_minv  #反归一化
y = y * (height_maxv - height_minv) + height_minv
plt.plot(show_labels.detach().numpy(), linestyle='-', color='red', linewidth=2, label='source')
plt.plot(y.detach().numpy(), linestyle='-', color='blue', linewidth=2, label='predict')
plt.legend()

error = y - show_labels
rmse = torch.mean(error ** 2) ** 0.5
print("rmse:", rmse.item())
```
