```python
import pywt
import numpy as np
import pandas as pd
from scipy.interpolate import make_interp_spline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('E:\实验数据\python\LSTM_data\data.csv', encoding='gbk')
load = df[df.columns[2]]
load = load.tolist()
```

```python
cA, cD = pywt.dwt(load, 'haar') #cA 是相似系数低频成分, cD 是细节系数高频成分
```

```python
x = [i for i in range(1, len(cA)+1)]
x_smooth = np.linspace(np.min(x), np.max(x), 600)
y_smooth = make_interp_spline(x, cA)(x_smooth)
plt.subplot(3,1,1)
plt.plot(x_smooth, y_smooth, c='green', marker='*', ms=1, alpha=0.75, label='true')

plt.subplot(3,1,2)
y_smooth = make_interp_spline(x, cD)(x_smooth)
plt.plot(x_smooth, y_smooth, c='red', marker='o', ms=1, alpha=0.75, label='pred')

plt.subplot(3,1,3)
x = [i for i in range(1, len(load)+1)]
x_smooth = np.linspace(np.min(x), np.max(x), 600)
y_smooth = make_interp_spline(x, load)(x_smooth)
plt.plot(x_smooth, y_smooth, c='red', marker='o', ms=1, alpha=0.75, label='pred')
# plt.grid(axis='y')
# plt.legend()
plt.show()
```
